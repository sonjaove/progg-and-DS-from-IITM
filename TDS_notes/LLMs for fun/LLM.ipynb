````python
#for training the model we can write with just one example.
def get_sentiment_other(review,smaple_Acting,sample_story,sampel_dir ,debug=False):
    response = requests.post(
        "http://aiproxy.sanand.workers.dev/openai/v1/chat/completions",
        headers={"Authorization": f"Bearer {api_key.api_token}"},
        json={
            "model": "gpt-3.5-turbo",
            "messages": [
                {"role": "system", "content": "Identify these based on the review:\nacting:\nstoryline:\ndirection:"},
                {"role": "user", "content": review},
                {"role": "assisstant", "content": f"acting: {smaple_Acting}\nstoryline: {sample_story}\ndirection: {sampel_dir}"},
                {"role": "user", "content": review}
            ]
        }
    )
    result = response.json()
    #make a seprate list of answers and then append it to the df if results is your original dataframe.
    answer = result["choices"][0]["message"]["content"]
    return answer
````
