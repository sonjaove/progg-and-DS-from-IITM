## the types of data.
1. public data 
    - open and free can be easily found on the ***www***

    - a particualr type of data is hard to find due to the vastness of it.

2. private data
    -  this kind of data is not open to everyone though it can be accessible to a group of people within the respective organization.

    - generally paid.

3. personal data
    - the data that specifically belongs to you, ex your call logs or your playlist. (though privacy is just a myth, but hypothtically sure!!)

    - personal logging as well.

## types of datasets
1. Structures dataset
    - no ambiguity and uncertanity

    - have a well defined schema i.e a 
predefined order

    - eg 
        - databases
        - spreadsheet (they're not always structured data)
        - shapefiles
2. Semi structured dataset
    - a json file - it has fields and values, but they are flexible.
    
    - documents 

        - eg
            - HTML
            - PDFs (looks structured but schema may not be easily extractible)
    
    - messages eg email and SMSs.

    - Containers 
       - eg 
            - ZIP
            - DOCX
            
3. Unstructured dataset
    - there is ambiguity and uncertanity

    - do not have a well defined schema.

    - eg
        - text
        - images
        - audio
        - video

## types of values
1. Categorical 
    - boolean
    - unordered - i.e cannot as such define a sequence or order
        - eg 
            - colors
            - gender
    - ordered - a sequence or order can be set up 
        - eg
            - high, medium, low 
            - small, medium, large
            - good, bad 
        - even a cyclical order is also possible 
            - eg
                - days of the week (while you cannot say which one is higher or lower but they definately form a cycle)
                - months
    - unstructured 
        - eg
            - text
            - binary

    - its possible to derive values form categorical data as these cols themselves don't have much operation to perform

2. Numerical 

    - reals and its subtets

3. Composite

    - Date/time
    - spatial 
        - eg
            - lat/lon
            - shapefiles
    - structured 
        - eg
            - json 
            - xml

    - specialized 
        - eg
            - IP 
            - currency
