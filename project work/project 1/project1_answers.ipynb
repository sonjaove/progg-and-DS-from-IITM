## 13. For the  given AC_NAME, What is the average vote share of the winners across all elections? Give your answer to 2 decimal places? [float]


vote share is votes of the candidate divided by the total votes in that election year expressed as a percentage

```python
def avg_vote_share(df_ac):
    #for getting the total votes for each year.
    total_votes = df_ac.groupby('YEAR')['VOTES'].sum().to_dict()
    #getting the votes of winning candidate for each year
    winnig_cadidate_votes = df_ac.groupby('YEAR')['VOTES'].max().to_dict()
    #vote share of winning candidate
    winnig_cadidate_votes_series = pd.Series(winnig_cadidate_votes)
    total_votes_series = pd.Series(total_votes)
    vote_share = winnig_cadidate_votes_series / total_votes_series
    avg_vote_share = vote_share.mean().round(2)
    return avg_vote_share
```

```python
#i assume i have to check the difference betweent the first and the last candidate for each year and see the difference and report the max difference.
def max_diff(df_ac):
    #getting the votes of winning candidate for each year
    winnig_cadidate_votes = df_ac.groupby('YEAR')['VOTES'].max().to_dict()
    #getting the votes of loosing candidate for each year
    loosing_cadidate_votes = df_ac.groupby('YEAR')['VOTES'].min().to_dict()
    #vote share of winning candidate
    winnig_cadidate_votes_series = pd.Series(winnig_cadidate_votes)
    loosing_cadidate_votes_series = pd.Series(loosing_cadidate_votes)
    diff = winnig_cadidate_votes_series - loosing_cadidate_votes_series
    max_diff = diff.max()
    return max_diff
```

```python
# creating a cache file, so that it runs faster when we try to re run the program.
def get_lat_lon(place, cache):
    if place in cache:
        return cache[place]
    location = geolocator.geocode(place)
    if location:
        cache[place] = (location.latitude, location.longitude)
        return location.latitude, location.longitude
    else:
        cache[place] = (None, None)
        return None, None

# Load cache if it exists
cache_file = 'kerela.pkl'
if os.path.exists(cache_file):
    cache = joblib.load(cache_file)
else:
    cache = {}

lat_lon_list = []
for place in tqdm(df['AC'], desc="Geocoding places"):
    lat_lon_list.append(get_lat_lon(place, cache))

df['lat'], df['lon'] = zip(*lat_lon_list)

# Save cache
joblib.dump(cache, cache_file)

# Drop rows with NaN values in lat or lon
df.dropna(subset=['lat', 'lon'], inplace=True,ignore_index=True)
```

```python
def group_constituency_data(df):
    # Group by 'AC' and find the row with max votes for each group
    df_grouped = df.sort_values('VOTES', ascending=False).groupby('AC').first().reset_index()
    
    # Select only the columns we need
    df_grouped = df_grouped[['AC', 'PARTY','SEX' ,'VOTES', 'lat', 'lon']]
    
    return df_grouped
```

```python
def count_constituencies_within_100km(df):
    # Initialize the count
    count = 0
    
    # Iterate over the rows in the DataFrame
    for i, row in df.iterrows():
        # Create a tuple with the latitude and longitude
        loc = (row['lat'], row['lon'])
        party=row['PARTY']
        
        # Iterate over the rows again to compare distances
        for j, row2 in df.iterrows():
            temp_party=row2['PARTY']
            # Skip comparison with itself
            if i == j:
                continue
            
            # Create a tuple with the latitude and longitude
            loc2 = (row2['lat'], row2['lon'])
            
            # Calculate the distance
            distance = geodesic(loc, loc2).km
            
            # Check if the distance is less than 100 km
            if distance < 100 and temp_party==party:
                count += 1
                break  # Once within 100 km, break out of the inner loop
    
    return count
```

## 17. Find the Pearson correlation coefficient between the votes won by female candidates and male candidates. Only include constituencies and election years that had at least 1 female candidate. Each row in the correlation data table should represent an election. Round off the answer to 4 decimal points[float]


```python
df = df[~(df['CANDIDATE'] == 'NOTA')]
df=df.reset_index(drop=True)
```

```python
def group_gender(df):
    # Dictionary to hold votes by gender for each (year, constituency) pair
    votes = {'F': {}, 'M': {}}
    
    # Iterate through each (year, constituency) pair
    for (year, ac), group in df.groupby(['YEAR', 'AC']):
        # Check if there is at least one female candidate in this group
        if (group['SEX'] == 'F').any():
            # Store the votes for female candidates
            female_votes = group[group['SEX'] == 'F']['VOTES'].tolist()
            if female_votes:
                votes['F'][(year, ac)] = sum(female_votes)
            
            # Store the votes for male candidates
            male_votes = group[group['SEX'] == 'M']['VOTES'].tolist()
            if male_votes:
                votes['M'][(year, ac)] = sum(male_votes)
    
    # Convert dictionary to DataFrame
    df_female_votes = pd.Series(votes['F']).to_frame(name='Female Votes')
    df_male_votes = pd.Series(votes['M']).to_frame(name='Male Votes')
    
    # Merge both DataFrames on the (year, constituency) index
    merged_df = df_female_votes.join(df_male_votes)
    
    # Drop rows with NaN values (where either female or male votes are missing)
    merged_df = merged_df.dropna()
    
    # Calculate Pearson correlation coefficient
    if not merged_df.empty:
        correlation, _ = pearsonr(merged_df['Female Votes'], merged_df['Male Votes'])
        return round(correlation, 4)
    else:
        return 0.0  # Return 0 if no valid correlation can be computed

```

## 18.. For a given state, identify the election year which has the most outliers on the basis of candidate votes. Use the interquartile range (IQR) rule to detect outliers. [integer]

```python
def find_year_with_most_outliers(df):
    # Filter data for the given state
    #df_state = df[df['STATE'] == state]
    
    # Dictionary to hold outlier counts for each year
    outlier_counts = {}
    
    # Iterate through each election year
    for year, group in df.groupby('YEAR'):
        # Calculate the IQR
        Q1 = group['VOTES'].quantile(0.25)
        Q3 = group['VOTES'].quantile(0.75)
        IQR = Q3 - Q1
        
        # Determine outlier boundaries
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Count outliers
        outliers = group[(group['VOTES'] < lower_bound) | (group['VOTES'] > upper_bound)]
        outlier_count = outliers.shape[0]
        
        # Store the count of outliers
        outlier_counts[year] = outlier_count
    
    # Identify the year with the most outliers
    max_outliers_year = max(outlier_counts, key=outlier_counts.get)
    
    return max_outliers_year
```

```python
print(find_year_with_most_outliers(df))
```
