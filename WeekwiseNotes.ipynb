```python
#sorting recurssively.
def mini(l):
    mini=l[0]
    for i in l:
        if i<mini:
            mini=i
    return mini
def sort(l):
    if l==[] or len(l)==1:
        return l
    m=mini(l)
    l.remove(m) # Shortning the list after every check.
    return[m]+sort(l) 
l=str(input('enter the list to be sorted.'))
nos=l.split(',')
lst=[int(num) for num in nos]
print(sort(lst))
```

since the above code increases linealry with the length of the list, after certain number the code wil take a loooooooooong time to complete the process.

therefore an effective method is to perform binary search, as it cuts down the number of iteraitons each time it exists the while loop, ***but for that the list has to be sorted***

```python
def binary_rec(l, k, begin, end):
    # beginning with a trivial case.
    if begin == end:
        return l[begin] == k
    # beginning the 'while' part of the binary search 
    if end - begin == 1:
        return l[begin] == k or l[end] == k
    if end - begin > 1:
        mid = (end + begin) // 2
        if l[mid] == k:
            return True
        elif l[mid] > k:
            return binary_rec(l, k, begin, mid - 1)  # Pass adjusted 'begin' and 'mid - 1'
        else:
            return binary_rec(l, k, mid + 1, end)   # Pass adjusted 'mid + 1' and 'end'
    return False
```

```python
# applying ceaser cypher on a file.
#k is the number of letters to be shifted.
import string
def ceaser_cypher(k):
    l=list(string.ascii_lowercase)
    d={}
    for i in range(len(l)):
        d[l[i]]=l[(i+k)%26]
    return d
```

```python
#reading a csv file.
def improvement(filename):
    # Open the file
    file = open(filename, 'r')
    # Read the first line (header) and discard it
    file.readline()
    # Initialize counter for students with increasing scores
    increasing_count = 0
    # Iterate over each line in the file
    for line in file:
        # Split the line into fields
        fields = line.strip().split(',')
        # Extract scores for CT, Python, and PDSA
        ct_score = int(fields[2])
        python_score = int(fields[3])
        pdsa_score = int(fields[4])
        # Check if scores are in the increasing order
        if ct_score < python_score < pdsa_score:
            increasing_count += 1
    # Close the file
    file.close()
    # Return the count of students with increasing scores
    return increasing_count
```

```python
def number_grid(m, n):
    file = open('numgrid.csv', 'w')
    start_number = 1
    for i in range(m):
        row = ""
        for j in range(n):
            row += str(start_number)
            if j != n - 1:
                row += ","
            start_number += 1
        file.write(row + '\n')
    file.close()
```

- ## Inheritance and Method overriding.
    - Inheritance is a fundamental concept that allows a class (subclass or child class) to inherit attributes and methods from another class (superclass or parent class). This enables code reuse and promotes a hierarchical structure among classes.
    - Method overriding is a feature in object-oriented programming that allows a subclass to provide a specific implementation of a method that is already defined in its superclass. When a subclass overrides a method, it provides its own implementation of the method with the same signature as the method in the superclass.
    -   - **Inheritance**: Establishes a relationship between classes, allowing subclasses to inherit attributes and methods from a superclass,  promoting code reuse and creating an "is-a" relationship between classes.
        - **Method Overriding**: Technique within inheritance where a subclass provides its own implementation for a method inherited from the superclass, altering the behavior of that method in the subclass context. It occurs within a single class hierarchy and allows for selective customization of inherited methods.
    -   - **Public Variables**: Accessible from outside the class, exposing internal state.
        - **Private Variables**: Accessible only within the class, encapsulating data to control access through public methods.
    - [Using import to perform inheritance and 'privacy' of a class](https://youtu.be/RrFvcDjJptY?feature=shared&t=1132)
    - ### Types of inheritance:
        1. **Single Inheritance**: Subclass inherits from only one superclass.
        
        2. **Multiple Inheritance**: Subclass inherits from more than one superclass, potentially leading to the diamond problem.
        
        3. **Multilevel Inheritance**: Subclass inherits from another subclass, creating a hierarchy.
        
        4. **Hierarchical Inheritance**: Multiple subclasses inherit from a single superclass, forming a hierarchy.
        
        5. **Hybrid (Virtual) Inheritance**: Combination of multiple and hierarchical inheritance, allowing a subclass to inherit from multiple superclasses.
